```python
import requests
from bs4 import BeautifulSoup

#grab the html data from the page
data_html=requests.get("https://www.amazon.in/b?ie=UTF8&node=7198569031").content
```

```python

bs=BeautifulSoup(data_html,"lxml")
```

```python

rows=bs.find_all("div",{"class":"vxd-music-bs-row"})
```

```python
len(rows)
```

```python
#in every row we are having 3 columns
cols=rows[0].find_all("div",{"class":"vxd-music-bs-info"})
```

```python
len(cols)
```

```python
cols[0].find("a").text.strip()
```

```python
cols[0].find("div",{"class":"a-row a-color-secondary a-size-small"}).text
```

```python
cols[0].find("span",{"class":"a-color-price"}).text.split("\xa0")[-1]
```

```python
fin=""
for r in rows:
    for c in r.find_all("div",{"class":"vxd-music-bs-info"}):
        a=c.find("a").text.strip().replace(",",":")
        b=c.find("div",{"class":"a-row a-color-secondary a-size-small"}).text
        v=c.find("span",{"class":"a-color-price"}).text.split("\xa0")[-1].replace(",","")
        fin+=a+","+b+","+v+"\n"
        print(a+","+b+","+v+"\n")
#         print(a)
```

```python
with open("datac.csv","w") as p:
    p.write(fin)
```
